```python
import os
import pandas as pd

def modify_timeseries_interval(interval_minutes):
    # Validate input
    if interval_minutes > 60 or 60 % interval_minutes != 0:
        print("Invalid input. The interval must be less or equal than 60 minutes and divide evenly into 60.")
        return

    # Get the current directory
    current_directory = os.getcwd()

    # Get the list of CSV files in the current directory
    scenario_files = [file for file in os.listdir(current_directory) if file.endswith('.csv')]

    # Read the first CSV file to determine the original time interval
    first_file = scenario_files[0]
    df_first = pd.read_csv(first_file, header=None)
    original_interval = len(df_first) // 8760

    if original_interval == 0:
        print("Invalid input. The original data does not have hourly values.")
        return

    # Loop through each scenario file and modify the interval
    for scenario_file in scenario_files:
        df_scenario = pd.read_csv(scenario_file, header=None)

        # Convert to hourly interval (delete repeated values within each hour)
        df_hourly = df_scenario.iloc[::original_interval]

        # Calculate the number of times to repeat the value
        num_repeats = 60 // interval_minutes

        # Repeat the value to match the desired interval
        df_modified = pd.DataFrame({'value': df_hourly[0].repeat(num_repeats)})

        # Save the modified data to the original CSV file (overwrite)
        df_modified.to_csv(scenario_file, header=False, index=False)

# Test the function with an interval of 30 minutes
interval_minutes = 60 # Change this value as needed
modify_timeseries_interval(interval_minutes)


```
